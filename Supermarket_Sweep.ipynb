a) For each pair of items $i, j \in \{1,...,56\}$, compute the shortest time $d_{ij}$ to go from item i to item j, and also store these values. Compute also the shortest time to travel between the start/end location and any item

```python
l = np.zeros((57,57))
cord = df.iloc[:,2:4]
cord.rename(columns={"X Position": 0, "Y Position": 1}, inplace = True)

for i in range(57):
    for j in range(57):
        if cord.loc[i,0] == cord.loc[j,0]:
            l[i,j] = abs(cord.loc[i,1]-cord.loc[j,1])
        else:
            dist_x = abs(cord.loc[i,0]-cord.loc[j,0])
            l[i,j] = min(110 - cord.loc[i,1] + dist_x + 110 - cord.loc[j,1], cord.loc[i,1] + dist_x + cord.loc[j,1])

# DataFrame with the shortest time dij to go from item i to item j
d = pd.DataFrame(l/10);d
```

```python
# Data
I=np.array(range(0,57))
J=np.array(range(0,57))
p = df['Price']

# Model
m = Model("Project")

# Decision Variable
x=m.addVars(I, J, name = 'x', vtype=GRB.BINARY)

#Objective
m.setObjective(quicksum(x[i,j]*p[j] for i in I for j in J),\
               sense = GRB.MAXIMIZE)
    

#Constraints
m.addConstrs((quicksum(x[i,j] for j in J) <= 1 for i in I),name='c1')
m.addConstrs((quicksum(x[i,j] for i in I) <= 1 for j in J),name='c2')
m.addConstrs((quicksum(x[j, k] for k in J[np.where(J != j)]) >= x[i,j] for i in I for j in J),name='c3')
m.addConstrs((quicksum(x[k, i] for k in I) >= x[i,j] for i in I for j in J),name='c4')
m.addConstr((quicksum(x[i,j] for i in I for j in J) <= 16),name='c5')
m.addConstr((quicksum(x[0,j] for j in J[1:]) == 1),name='c6')
m.addConstr((quicksum(x[i,0] for i in I[1:]) == 1),name='c7')
m.addConstr((quicksum((d.iloc[i,j]+2)*x[i,j] for i in I for j in J)-2 <= 90),name='c8')
m.addConstr((quicksum(x[i,i] for i in I) == 0),name='c9')
    
m.optimize()

opt = pd.DataFrame(np.zeros((57,57)))
print("Optimal solution:")
for v in m.getVars():
    x = int(v.varName[2:-1].split(',')[0])
    y = int(v.varName[2:-1].split(',')[1])
    opt.iloc[x,y] = v.x
    if v.x == 1:
        print("%s = %g" % (v.varName, v.x))

print("Optimal objective value:\n{}".format(m.objVal))
```
